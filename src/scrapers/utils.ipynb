```python
import csv
import os
from datetime import datetime, timezone
from typing import List, Dict
```

```python
def save_csv(rows: List[Dict], filename: str):
    if not rows:
        return

    os.makedirs(os.path.dirname(filename), exist_ok=True)

    with open(filename, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=rows[0].keys())
        writer.writeheader()
        writer.writerows(rows)

    print(f"Saved {len(rows):,} rows to {filename}")
```

```python
def timestamped_filename(base_name: str, directory: str, extension: str = "csv") -> str:
    """
    Build a timestamped output filename.
    e.g. timestamped_filename("polymarket", "data/raw") →
         "data/raw/polymarket_20260315_143022.csv"
    """
    ts = datetime.now(timezone.utc).strftime("%Y%m%d_%H%M%S")
    return os.path.join(directory, f"{base_name}_{ts}.{extension}")
```
